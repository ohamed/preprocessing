### Reading the data using read_pickle() + data exploration!

```python
# Read the cases dataframe 

cases_df = pd.read_pickle(r"..\data\df_cases_200906.gzip")
cases_df.shape
```

```python
# Show the top five rows in df1!

cases_df.head()
```

```python
# Read the labels dataframe 

labels_df = pd.read_pickle(r'..\data\df_label_200906.gzip')
labels_df.shape
```

```python
# Show the top five rows in df2!

labels_df.head()
```

### Begin of Data_Preparation

```python
#Joining the two dataframes into one, based on CaseId!

df_final = pd.merge(cases_df, labels_df, on=['CaseId'])
df_final.shape
```

```python
# Show the top five rows in df_final after merge!

df_final.head()
```

#### To get rid of some uneeded columns: Language, StartDate and DocumentType!

```python
#delete Language
if 'Language' in df_final.columns:
    del df_final['Language']
```

```python
#delete StartDate
if 'StartDate' in df_final.columns:
    del df_final['StartDate']
```

```python
#delete DocumentType
if 'DocumentType' in df_final.columns:
    del df_final['DocumentType']
```

```python
df_final.head(10)
```

```python
#check the condition, to be used by np.select()

[ (df_final['IsExecuted'] == False) | (df_final['QualityScore'] <0.81) ]
```

```python
# ===== Add an INVALID, column, which represents INVALID Contracts  =====
#- `IsExecuted==False`
#- `QualityScore<0.81

# create a list of conditions!
conditions = [ (df_final['IsExecuted'] == False) | (df_final['QualityScore'] <0.81) ]


df_final['INVALID'] = np.select(conditions, "1", "0")
```

```python
df_final.head(10)
```

```python
## As per the requirments in README.md file: 
## 1- Get rid of OcrText for INVALID Contracts!

df_final.loc[df_final.INVALID == "1", "OcrText"] = ""

## 2- Add two new columns to specific index: ValidFileNames and InalidFileNames
df_final.insert(2, 'InalidFileNames', "")
df_final.insert(3, 'ValidFileNames', "")
```

```python
print(df_final.shape)
df_final.head(10)
```

```python
## As per the "df_final_eg.png", contracts file names should be in InalidFileNames and ValidFileNames!

df_final.loc[df_final.INVALID == "1", "InalidFileNames"] = df_final.FileName
df_final.loc[df_final.INVALID == "0", "ValidFileNames"] = df_final.FileName
```

```python
df_final.head(10)
```

#### To get rid of some additional uneeded columns: IsExecuted, QualityScore and INVALID!

```python
#delete IsExecuted
if 'IsExecuted' in df_final.columns:
    del df_final['IsExecuted']
```

```python
#delete QualityScore
if 'QualityScore' in df_final.columns:
    del df_final['QualityScore']
```

```python
#delete INVALID
if 'INVALID' in df_final.columns:
    del df_final['INVALID']
```

```python
#delete FileName
if 'FileName' in df_final.columns:
    del df_final['FileName']
```

```python
print(df_final.shape)
df_final.head()
```

```python
## Get a copy of df_final dataframe!

df_to_save = df_final.copy()
df_to_save.head()
```

```python
## Use dataframe groupby with aggregation_functions!

#aggregation_functions = { 'InalidFileNames': ','.join, 'ValidFileNames': ','.join, 'OcrText':' '.join, 'label_1':'first', 'label_2':'first'}
aggregation_functions = { 'InalidFileNames':list, 'ValidFileNames': list, 'OcrText':' '.join, 'label_1':'first', 'label_2':'first'}

df_to_save = df_to_save.groupby(df_to_save['CaseId'], as_index=False).aggregate(aggregation_functions)
```
